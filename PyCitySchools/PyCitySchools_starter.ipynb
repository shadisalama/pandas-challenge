```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_count = len(school_data_complete["school_name"].unique())
#school_count
studen_count = school_data_complete["Student ID"].count()
#studen_count
total_budget = school_data["budget"].sum()
#total_budget
avg_math_score = school_data_complete["math_score"].mean()
#avg_math_score
avg_reading_score = school_data_complete["reading_score"].mean()
#avg_reading_score
overall_avg_score = (avg_math_score + avg_reading_score)/2
#overall_avg_score
math_passing_score = school_data_complete[school_data_complete["math_score"]>=70].count()["student_name"]
math_percentage = (math_passing_score/studen_count)*100
#math_percentage
reading_passing_score = school_data_complete[school_data_complete["reading_score"]>=70].count()["student_name"]
reading_percentage = (reading_passing_score/studen_count)*100
reading_percentage
```

```python
summary = pd.DataFrame({"total school":school_count,
                       "total students":studen_count,
                       "total budget":[total_budget],
                       "average math score":[avg_math_score],
                       "average reading score":[avg_reading_score],
                       "overall average score":[overall_avg_score],
                       "% math passing percentage":[math_percentage],
                       "% reading passing percentage":[reading_percentage]})

summary["total students"] = summary["total students"].map("{:,}".format)
summary["total budget"] = summary["total budget"].map("${:,.2f}".format)
summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_type = school_data.set_index(["school_name"])["type"]
#school_type
school_grouped = school_data_complete.groupby(["school_name"])
student_per_school = school_grouped["student_name"].count()
#student_per_school
budget_per_school = school_grouped["budget"].mean()
#budget_per_school
per_student_budget = budget_per_school/student_per_school
#per_student_budget
avg_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
#avg_school_math
avg_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
#avg_school_reading
passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]

school_passing_math = passing_math.groupby(["school_name"]).count()["student_name"] / student_per_school * 100
school_passing_reading = passing_reading.groupby(["school_name"]).count()["student_name"] / student_per_school * 100
overall_passing_rate = (school_passing_math + school_passing_reading) / 2
#overall_passing_rate

```

```python
school_summary = pd.DataFrame({"school type": school_type,
                                   "total students": student_per_school,
                                   "total school budget": budget_per_school,
                                   "per student budget": per_student_budget,
                                   "average math score": avg_school_math,
                                   "average reading score": avg_school_reading,
                                   "% passing math": school_passing_math,
                                   "% passing reading": school_passing_reading,
                                   "% overall passing rate": overall_passing_rate})


school_summary = school_summary[["school type", "total students", "total school budget", "per student budget",
                                         "average math score", "average reading score", 
                                         "% passing math", "% passing reading", 
                                         "% overall passing rate"]]
school_summary["total school budget"] = school_summary["total school budget"].map("${:,.2f}".format)
school_summary["per student budget"] = school_summary["per student budget"].map("${:,.2f}".format)


school_summary
```
